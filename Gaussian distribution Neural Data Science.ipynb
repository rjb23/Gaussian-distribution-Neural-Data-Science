```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate random data following a Gaussian distribution
mean = 0  # Mean of the distribution
std = 1   # Standard deviation of the distribution
size = 1000  # Number of data points to generate
data = np.random.normal(mean, std, size)

# Calculate the mean and standard deviation of the generated data
data_mean = np.mean(data)
data_std = np.std(data)

# Plot a histogram of the generated data
plt.hist(data, bins='auto', density=True, alpha=0.7, rwidth=0.85)

# Plot the probability density function (PDF) of the Gaussian distribution
x = np.linspace(data_mean - 3*data_std, data_mean + 3*data_std, 100)
plt.plot(x, norm.pdf(x, data_mean, data_std), 'r-', lw=2)

# Add labels and title to the plot
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Gaussian Distribution')

# Show the plot
plt.show()
```

```python
#In this example, we generate random data following a Gaussian distribution using the np.random.normal function. We then calculate the mean and standard deviation of the generated data using NumPy's np.mean and np.std functions. Next, we plot a histogram of the data using Matplotlib's plt.hist function. Finally, we plot the probability density function (PDF) of the Gaussian distribution using SciPy's norm.pdf function.

#By adjusting the mean, std, and size parameters, you can generate data with different characteristics and explore the properties of the Gaussian distribution.

#Note that the above example assumes you have the necessary libraries (numpy, matplotlib, scipy) installed in your Python environment. You can install them using pip install numpy matplotlib scipy.
```
